```python
import numpy as np
import scipy.stats as stats
```

```python
# Comparing the performance of two versions of algorithm If the new one is usefull or not.

# 1. The Data Simulating User Level Group Data
# control group : 50$ avg spend, 20$ std dev, 500 users.

control_group = np.random.normal(loc=50, scale=20, size=500)

# Results produced by new ml model.
# Variant Group : 53$ avg spend, 25$ std dev, 500 users.

variant_group = np.random.normal(loc=53, scale=25, size = 500)

t_stat, p_value = stats.ttest_ind(control_group, variant_group, equal_var=False)

print(f"Average Control Group:{np.mean(control_group):.2f}")
print(f"Average Variant Group:{np.mean(variant_group):.2f}")
print(f"P-Value:{p_value:.4f}")
```

```python
if p_value < 0.05:
    print("Reject Null : The revenue increase is statistically significant.")
else:
    print("Failed to reject null hypothesis: Its just Noise.")
```
