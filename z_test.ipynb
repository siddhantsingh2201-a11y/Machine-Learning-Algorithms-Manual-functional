```python
import pandas as pd
import numpy as np
import statsmodels.stats.proportion as proportion 
```

```python
conversions = np.array([1200,1300])
samples = np.array([10000,10000])
```

```python
# 2. The Test
# Value = 0 means null hypothesis is difference = 0.
stat, p_value = proportion.proportions_ztest(conversions, samples, value=0, alternative="two-sided")
alpha = 0.05

print(f"Z-score: {stat:.4f}")
print(f"P_Value: {p_value:.4f}")
```

```python
if p_value < alpha:
    print("reject null hypothesis")
else:
    print("failed to reject the null hypothesis.")
```
